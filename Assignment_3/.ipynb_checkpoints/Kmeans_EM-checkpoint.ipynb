```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

#K Means function... TAKES ABOUT 40 SECONDS.
def km(img, K, iterations):
    centroids = []
    
    #create random centroids
    for i in range(K):
        centroids.append(np.random.randint(0, 256, 3))
    centroids = np.array(centroids)

    for i in range(iterations):
    
        if(3 == K):
             #First find ecluadian distance between each pixel and the centroids, then take the index
             #of the smallest distance
            labels = np.array([np.argmin([np.linalg.norm(j-centroids[0]),np.linalg.norm(j-centroids[1]), np.linalg.norm(j-centroids[2])], axis=0)for j in img])
        else:
            #same as in the previous one, but with 2 centroids instead of 3
            labels = np.array([np.argmin([np.linalg.norm(j-centroids[0]),np.linalg.norm(j-centroids[1])], axis=0)for j in img])

        #recreate the centroids based on previous labels
        centroids = []
        for n in range(K):
            f = img[labels == n]
            centroids.append(np.average(f, axis=0))

    return labels, centroids

img = cv2.imread('home.jpg')
x = img.reshape((-1,3))
height, width = img.shape[:2]
l2, c2 = km(x, 2, 10)
l3, c3 = km(x, 3, 10)

im2 = np.reshape(l2, (height, width))
im3 = np.reshape(l3, (height, width))

plt.figure(figsize=(15,15))
plt.subplot(221), plt.imshow(im2)
plt.title('K = 2'), plt.xticks([]), plt.yticks([])
plt.subplot(222), plt.imshow(im3)
plt.title('K = 3'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#Expectation Maximaization

# function to calculate the variance of a given image
def calculate_gaussian(x, mean, var):
    #Gaussian gaussian distribution equation
    return (1 / (np.sqrt(2*np.pi) * np.sqrt(var))) * np.exp(-((x - mean)**2) / (2*var))
    
#function to compress an array to 1/3 of its size (to be used for the Labels)
def compress(arr):
    final = []
    i = 0
    while i < (len(arr)-2):
        final.append(int(np.median(arr[i:i+3], axis=0)))
        i += 3
    return np.array(final)
    
    
# function to find the labels and centroids of an image using Expectation maximaization
def EM(img, K, iterations):
    #flatten image, convert to one long array
    img = img.flatten()
    
    #Some of the logic below was inspired by:
    #https://github.com/gcorcorann/ECSE415/blob/master/Assign3/Assignment3.ipynb
    
    #random clusters
    prev_idx = 0
    labels = np.zeros_like(img)
    for k in range(K):
        idx = (1/K) * 255 * (k+1)
        labels[np.where((img >= prev_idx) & (img < idx))] = k
        prev_idx = idx
    
    for i in range(iterations):
        
        #Maximaization part
        #start with maximaization because we already have a random cluster

        mu = [] #updated means
        var = [] #updated variances
        alpha = [] #gaussian weighing factors
        for k in range(K):
            X = img[np.where(labels==k)[0]]
            mu.append(np.mean(X))
            var.append(np.var(X))
            alpha.append(len(X) / len(img))

        #Expectation part
        probs = np.array([calculate_gaussian(img, mu[n], var[n]) for n in range(K)]).T
        probs *= alpha
        for row in probs:
            row /= sum(row)
        labels = np.argmax(probs, axis=1)
        
    return mu, labels

img = cv2.imread('home.jpg')
height, width = img.shape[:2]

#find labels and centroids using EM
centroids2gx, labels2gx = EM(img, 2, 10)
centroids3gx, labels3gx = EM(img, 3, 10)

#reshape labels to the shape of the image
img_label2gx = np.reshape(compress(labels2gx), (height, width))
img_label3gx = np.reshape(compress(labels3gx), (height, width))


# display images
plt.figure(figsize=(15,15))
plt.subplot(221), plt.imshow(img_label2gx)
plt.title('K = 2'), plt.xticks([]), plt.yticks([])
plt.subplot(222), plt.imshow(img_label3gx)
plt.title('K = 3'), plt.xticks([]), plt.yticks([])
plt.show()
```

#### Under what data distribution conditions would K-means and EM give the same solution, provided similar initializations?

Some of k-means features are: 
- assumes clusters are spherical
- hard assignment of elements (either in cluster A or B)

Therefore, k-means and EM would produce similar results if the clusters were actually spherical and reltivlay distant apart from each other (so that the hard assignment of k-means can correlate to the probability of EM)
