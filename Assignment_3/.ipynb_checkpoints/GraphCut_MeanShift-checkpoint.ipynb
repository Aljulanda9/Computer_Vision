```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from skimage import segmentation, color
from skimage.future import graph
from skimage.segmentation import quickshift

#GRAPH CUT SEGMENTATION
img = cv2.cvtColor(cv2.imread('home.jpg'), cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10,5))
plt.subplot(121), plt.imshow(img)
plt.title("Original Image"), plt.xticks([]), plt.yticks([])
plt.show()

def graph_cut(img, compactness, segments, threshold):
    labels1 = segmentation.slic(img, compactness=compactness, n_segments=segments)
    out1 = color.label2rgb(labels1, img, kind='avg')
    
    # apply graph cut
    g = graph.rag_mean_color(img, labels1, mode='similarity')
    labels2 = graph.cut_normalized(labels1, g, thresh=threshold)
    out2 = color.label2rgb(labels2, img, kind='avg')

    # display
    plt.figure(figsize=(10,5))
    plt.subplot(121), plt.imshow(out2)
    plt.title("compactness = " + str(compactness) + ", segments = " + str(segments) + ", thres = " + str(threshold))
    plt.xticks([]), plt.yticks([])
    plt.show()

print('=== MANIPULATE THRESHOLD ===')
graph_cut(img, 10, 300, 0.001)
graph_cut(img, 10, 300, 0.1)
graph_cut(img, 10, 300, 10)

print('\n\n=== MANIPULATE compactness ===')
graph_cut(img, 0.1, 300, 0.001)
graph_cut(img, 2, 300, 0.001)
graph_cut(img, 10, 300, 0.001)

print('\n\n=== MANIPULATE n_segments ===')
graph_cut(img, 10, 10, 0.001)
graph_cut(img, 10, 30, 0.001)
graph_cut(img, 10, 100, 0.001)
```

### Parameters effect:
###### compactness: 
Balances color proximity and space proximity. Higher values give more weight to space proximity, making superpixel shapes more square/cubic. 

###### n_segments: 
APPROXIMATE number of labels. More n_segments mean the pixels are classified into more groups/clusters.

###### threshold: 
A subgraph wont be further subdivided if the value of the N-cut exceeds thresh. Large threshold leaves a more superpixels 'unmerged\disconnected'. Low threshold 'merges\connects' more superpixels. 

```python
#MEAN SHIFT SEGMENTATION

#plot original image
plt.figure(figsize=(10,5))
plt.subplot(121), plt.imshow(img)
plt.title("Original Image"), plt.xticks([]), plt.yticks([])
plt.show()

def mean_shift(img, ratio, kernel, max_dist):
    label = quickshift(img, ratio = ratio, kernel_size = kernel, max_dist=max_dist)
    out = color.label2rgb(label, img, kind='avg')
    # display
    plt.figure(figsize=(10,5))
    plt.subplot(121), plt.imshow(out)
    plt.title("ratio = " + str(ratio) + ", kernel_size = " + str(kernel) + ", max_dist = " + str(max_dist))
    plt.xticks([]), plt.yticks([])
    plt.show()

    
print('=== MANIPULATE max_dist ===')
mean_shift(img, 1.0, 5, 5)
mean_shift(img, 1.0, 5, 20)
mean_shift(img, 1.0, 5, 150)

print('\n\n=== MANIPULATE kernel_size ===')
mean_shift(img, 1.0, 2, 10)
mean_shift(img, 1.0, 5, 10)
mean_shift(img, 1.0, 15, 10)

print('\n\n=== MANIPULATE ratio ===')
mean_shift(img, 0.1, 5, 10)
mean_shift(img, 0.5, 5, 10)
mean_shift(img, 1.0, 5, 10)
```

### Parameters effect:
###### ratio: 
Balances color-space proximity and image-space proximity. Higher values give more weight to color-space. 

###### kernel_size: 
Width of Gaussian kernel used in smoothing the sample density. Higher means fewer clusters.

###### max_dist: 
Cut-off point for data distances. Higher means fewer clusters.
