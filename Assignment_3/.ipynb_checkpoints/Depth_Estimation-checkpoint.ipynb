```python
import sys
import re
from struct import *
import numpy as np
from matplotlib import pyplot as plt

def read_pfm(file):
    # Adopted from https://stackoverflow.com/questions/37073108/how-to-read-pfm-files-provided-with-middlebury-dataset?rq=1
    with open(file, "rb") as f:
        # Line 1: PF=>RGB (3 channels), Pf=>Greyscale (1 channel)
        type = f.readline().decode('latin-1')
        if "PF" in type:
            channels = 3
        elif "Pf" in type:
            channels = 1
        else:
            sys.exit(1)
        # Line 2: width height
        line = f.readline().decode('latin-1')
        width, height = re.findall('\d+', line)
        width = int(width)
        height = int(height)

        # Line 3: +ve number means big endian, negative means little endian
        line = f.readline().decode('latin-1')
        BigEndian = True
        if "-" in line:
            BigEndian = False
        # Slurp all binary data
        samples = width * height * channels;
        buffer = f.read(samples * 4)
        # Unpack floats with appropriate endianness
        if BigEndian:
            fmt = ">"
        else:
            fmt = "<"
        fmt = fmt + str(samples) + "f"
        img = unpack(fmt, buffer)
        img = np.flipud(np.array(img).reshape(height, width))
        img[img==np.inf] = 0
    return img
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

def depth(imgL, imgR, disparities, block, ground):
    # compute disparity map
    stereo = cv2.StereoBM_create(numDisparities=disparities, blockSize=block)
    disparity = stereo.compute(imgL, imgR)
    
    # calculate pixel-wise squared difference
    pwd = (np.array(ground) - np.array(disparity))**2

   # display images
    plt.figure(figsize=(15,15))
    plt.subplot(121), plt.imshow(disparity, cmap="gray")
    title = "Disparity Map block = " + str(block) + " and disp = "+ str(disparities)
    plt.title(str(title)), plt.xticks([]), plt.yticks([])
    plt.subplot(122), plt.imshow(pwd, cmap="gray")
    plt.title("Pixel wise square difference with ground"), plt.xticks([]), plt.yticks([])
    plt.show()
    return disparity
    
# load left and right images
imgL = cv2.imread("im0.png", 0)
imgR = cv2.imread("im1.png", 0)

# resize images for fast processing
imgL = cv2.resize(imgL, None, fx=0.3, fy=0.3)
imgR = cv2.resize(imgR, None, fx=0.3, fy=0.3)

#pfm image
depth_img = read_pfm('disp0.pfm')
depth_img= cv2.resize(depth_img, (imgL.shape[1], imgL.shape[0]))


disparities = [16,32,48]
blocks = [5,9,15]

for disparity in disparities:
    for block in blocks:
        depth(imgL, imgR, disparity, block, depth_img)
```

#### Comment on the error-maps for each combination of parameters. Specifically, mention the effects of varying the parameters on the error in the region of high/low texture, edge, large/small depths.

Generally, increasing block size means losing details (or smoothening the map) and decreasing it might lead to having more noise effects (but preserving details).

Also, number of disparities is inversely proportional to the distance from the camera, meaning a higher disparity range allows us to reconstruct objects that are closer to the camera, while lower disparity range constructs more of the objects that are further away.

As to the error maps, the effects of increasing the number of disparities are as follows for the 3 specific regions:
- high/low texture: more texture can be seen as number of disparity increases.
- edges: edges become sharper and more visible.
- large/small depth: areas behind the bike become more visible i.e. more depth can be observed.

And the effects of increasing the block size are as follows for the 3 specific regions:
- high/low texture: as the image gets smoother with the increase in block size, texture becomes less visible.
- edges: The edges in the error image become smoother.
- large/small depth: can't observe an effect.
