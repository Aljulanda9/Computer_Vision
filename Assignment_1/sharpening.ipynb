```python
##############################
# ASSIGNMENT 1: Question 1.2 #
##############################

#1.2 part 1 sharpen images
import numpy as np
import cv2
import matplotlib.pyplot as plt

#image to be processed
rice = cv2.imread("./rice.png", cv2.IMREAD_GRAYSCALE)

#5*5 box filtered image
boxBlur = cv2.blur(rice, (5,5))

#5*5 gaussian filterd image
gaussBlur = cv2.GaussianBlur(rice, (5,5), np.sqrt(2))

#sharpening equation for box
boxSharpened = rice + rice - boxBlur

#sharpening equation for gaussian
gaussSharpened = rice + rice - gaussBlur

#plot
plt.figure(figsize=(30, 30))
plt.subplot(131), plt.imshow(rice, cmap="gray")
plt.title("Original"), plt.xticks([]), plt.yticks([])
plt.subplot(132), plt.imshow(boxSharpened, cmap="gray")
plt.title("boxSharpened"), plt.xticks([]), plt.yticks([])
plt.subplot(133), plt.imshow(gaussSharpened, cmap="gray")
plt.title("gaussSharpened"), plt.xticks([]), plt.yticks([])
plt.show()
```

After printing the matrix of the details image = original - blur, I found that (original - boxBlurred) has higher pixel values than (original - gaussinaBlurred), which means that the details obtained from using the boxBlur are larger. Therefore, the boxBlur should give better sharpeneing. 

This is expected as the kernel of the box filter is uniform (all values = 1/25) while the kernel of the gaussian 5*5 kernel has varying pixel values and many of which are less than (1/25).

However, the difference in pixel values of the details images obtained by gaussian and box filters was beteween "1 to 4" in terms of intenecities which explains why it is very hard to observe the difference just by looking at the images.
